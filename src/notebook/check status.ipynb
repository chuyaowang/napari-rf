```python
stacks = [
    '/media/philipp/WD1/clb2/data/20230511_CLB2cellcycle_fovs002.nd2',
    '/media/philipp/WD1/clb2/data/20230511_CLB2cellcycle_fovs003.nd2',
    '/media/philipp/WD1/clb2/data/20230512_CLB2cellcycle_fovs004.nd2']
```

```python
processed_data = [
  '/media/philipp/WD1/clb2/processed_data/20230511_CLB2cellcycle_fovs002/', 
  '/media/philipp/WD1/clb2/processed_data/20230511_CLB2cellcycle_fovs003/', 
  '/media/philipp/WD1/clb2/processed_data/20230512_CLB2cellcycle_fovs004/']
```

```python
mappings = {
    '/media/philipp/WD1/clb2/processed_data/20230511_CLB2cellcycle_fovs002/':
    {
        'wt': [5,6,7,8,9],
        'zip': [15, 16, 17,18,19],
        'she2': [20,21,22,23,24],
        'she3': [10, 11, 12,13,14],
    },
    '/media/philipp/WD1/clb2/processed_data/20230511_CLB2cellcycle_fovs003/':
    {
        'wt': [10,11,12,13,14],
        'zip': [15,16,17,18,19],
        'she2': [5,6,7,8,9],
        'she3': [20,21,22,23,24],
    },
    '/media/philipp/WD1/clb2/processed_data/20230512_CLB2cellcycle_fovs004/':
    {
        'wt': [5,6,7,8,9],
        'zip': [15,16,17,18,19],
        'she2': [20,21,22,23,24],
        'she3': [10,11,12,13,14],
    },
}
```

```python
for stack_path, path in zip(stacks, processed_data):
    stack = nd2(stack_path)
    strains = mappings.get(path)
    for strain, positions in strains.items():
        for position in positions:
            processed_path = f"{path}/position_{str(position).zfill(2)}/z_level_0/"
            img_paths = glob(f"{processed_path}/neck_segmentation/*")
            if stack.sizes['t'] != len(img_paths):
                present = []
                print('-'*20)
                print(path, strain, position)
                for im in img_paths:
                    matches = re.findall('frame_(\d+)', im)
                    if matches:
                        present.append(int(matches[0]))
                absent = [a for a in range(stack.sizes['t']) if not a in present]
                print(absent)
                print('-'*20)
```
